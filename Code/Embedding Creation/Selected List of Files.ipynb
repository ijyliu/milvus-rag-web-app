# Selected List of Files

 Limit list of files based off seleted list of companies

```python
# Packages
import pandas as pd
```

## Load data

```python
selected_list_of_companies = pd.read_excel('selected_list_of_companies.xlsx')
selected_list_of_companies
```

```python
# limit to where keep is 1
selected_list_of_companies = selected_list_of_companies[selected_list_of_companies['keep'] == 1]
# Rename column 0 to company
selected_list_of_companies.rename(columns = {0: 'company'}, inplace = True)
# Drop keep column
selected_list_of_companies.drop(columns = ['keep'], inplace = True)
selected_list_of_companies
```

```python
list_of_files = pd.read_excel('list_of_files.xlsx')
list_of_files
```

```python
# Rename 0 to filename
list_of_files.rename(columns = {0: 'filename'}, inplace = True)
# Get company as first part of filename before underscore
list_of_files['company'] = list_of_files['filename'].str.split('_').str[0]
list_of_files
```

## Perform selection via inner join

```python
selected_list_of_files = list_of_files[list_of_files['company'].isin(selected_list_of_companies['company'])][['filename']]

# Output to excel
selected_list_of_files.to_excel('selected_list_of_files.xlsx', index = False)

selected_list_of_files
```

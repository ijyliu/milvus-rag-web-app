```python
import requests
import json
```

```python
# Load service account credentials
import os
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = './Credentials/milvus-rag-web-app-service-account.json'
```

```python
# Load URL/URI from Credentials/query_embedding_model_uri.txt
with open('Credentials/query_embedding_model_uri.txt', 'r') as file:
    url = file.read().strip()
    url = url + '/api/embeddings'
```

```python
# Get ID token
import google.auth.transport.requests
import google.oauth2.id_token

auth_req = google.auth.transport.requests.Request()
id_token = google.oauth2.id_token.fetch_id_token(auth_req, audience=url)
```

```python
# Set headers and data for the request, and authentication
headers = {'Content-Type': 'application/json', 'Authorization': f'Bearer {id_token}'}
data = {"model": "mxbai-embed-large:335m",
        "prompt": "Represent this sentence for searching relevant passages: The sky is blue because of Rayleigh scattering"}
```

```python
response = requests.post(url, headers=headers, data=json.dumps(data))
```

```python
import requests
import json
```

```python
# Set headers and data for the request
headers = {'Content-Type': 'application/json'}
data = {"model": "mxbai-embed-large:335m",
        "prompt": "Represent this sentence for searching relevant passages: The sky is blue because of Rayleigh scattering"}
```

```python
response = requests.post('http://localhost:5000/api/embeddings',
                         headers=headers, 
                         data=json.dumps(data))
```

```python
print(response.status_code)
```

```python
print(response.content)
```

```python
print(response.text)
```

```python
print(response.json())
```

```python
print(response.json()['embedding'])
```
